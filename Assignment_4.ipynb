```python
from sentence_transformers import SentenceTransformer
import numpy as np

# تحميل المودل
model = SentenceTransformer("all-MiniLM-L6-v2")

# مجموعة النصوص
documents = [
    "Artificial intelligence is transforming healthcare by enabling early diagnosis.",
    "Machine learning helps detect diseases from medical images.",
    "E-commerce platforms use recommendation systems to improve sales.",
    "Autonomous vehicles rely on AI for navigation and safety.",
    "AI chatbots are used in customer service to answer user questions."
]

# تحويل النصوص إلى embeddings
doc_embeddings = model.encode(documents)

# دالة البحث الدلالي
def semantic_search(query, documents, embeddings):
    query_embedding = model.encode([query])[0]

    similarities = np.dot(embeddings, query_embedding) / (
        np.linalg.norm(embeddings, axis=1) * np.linalg.norm(query_embedding)
    )

    best_match_index = np.argmax(similarities)

    return documents[best_match_index], similarities[best_match_index]

# تجربة
query = "How is AI used in medicine?"
result, score = semantic_search(query, documents, doc_embeddings)

print("Query:", query)
print("Best Match:", result)
print("Similarity Score:", score)
```

```python
test_queries = [
    "How does AI help doctors?",
    "What is AI used for in customer support?",
    "How do online stores increase sales?",
    "How do self-driving cars work?"
]

for q in test_queries:
    result, score = semantic_search(q, documents, doc_embeddings)
    print("=" * 60)
    print("Query:", q)
    print("Best Match:", result)
    print("Similarity Score:", score)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

documents = [
    "Artificial intelligence is transforming healthcare by enabling early diagnosis.",
    "Machine learning helps detect diseases from medical images.",
    "E-commerce platforms use recommendation systems to improve sales.",
    "Autonomous vehicles rely on AI for navigation and safety.",
    "AI chatbots are used in customer service to answer user questions."
]

doc_embeddings = model.encode(documents)

def semantic_search_top_k(query, documents, embeddings, k=3):
    query_embedding = model.encode([query])[0]

    similarities = np.dot(embeddings, query_embedding) / (
        np.linalg.norm(embeddings, axis=1) * np.linalg.norm(query_embedding)
    )

    top_k_indices = np.argsort(similarities)[::-1][:k]

    results = []
    for idx in top_k_indices:
        results.append((documents[idx], similarities[idx]))

    return results

test_queries = [
    "How does AI help doctors?",
    "What is AI used for in customer support?",
    "How do online stores increase sales?",
    "How do self-driving cars work?"
]

print("Semantic Search Results:\n")

for q in test_queries:
    print("=" * 60)
    print("Query:", q)
    results = semantic_search_top_k(q, documents, doc_embeddings, k=3)

    for i, (doc, score) in enumerate(results, 1):
        print(f"Top {i} Match: {doc}")
        print(f"Similarity Score: {score:.4f}")
    print()
```
