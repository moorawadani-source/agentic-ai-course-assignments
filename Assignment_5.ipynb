```python
from google.colab import userdata
import os

os.environ["OPENROUTER_API_KEY"] = userdata.get("OPENROUTER_API_KEY").strip()
```

```python
documents = [
    "Artificial intelligence can personalize learning experiences for students.",
    "AI can automate grading and administrative tasks for teachers.",
    "AI in education raises concerns about privacy, bias, and academic integrity.",
    "AI tutoring systems can adapt to each student's learning pace."
]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=20
)

docs = text_splitter.create_documents(documents)
print(docs[:2])
```

```python
!pip install -q sentence-transformers
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_community.vectorstores import Chroma

vectorstore = Chroma.from_documents(
    documents=docs,
    embedding=embeddings
)
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model="nvidia/nemotron-3-nano-30b-a3b:free",
    temperature=0,
    base_url="https://openrouter.ai/api/v1",
    api_key=os.environ["OPENROUTER_API_KEY"],
)
```

```python
results = retriever.invoke("What are the risks of AI in education?")
for r in results:
    print(r.page_content)
    print("-" * 50)
```

```python
question = "What are the benefits and risks of AI in education?"

retrieved_docs = retriever.invoke(question)
context = "\n\n".join([doc.page_content for doc in retrieved_docs])

prompt = f"""
Answer the question using only the context below.

Context:
{context}

Question:
{question}
"""

response = model.invoke(prompt)
print(response.content)
```

```python
retrieved_docs = retriever.invoke(question)

print("Retrieved Documents:")
for i, doc in enumerate(retrieved_docs, 1):
    print(f"\nDocument {i}:")
    print(doc.page_content)
    print("-" * 50)

context = "\n\n".join([doc.page_content for doc in retrieved_docs])

prompt = f"""
Answer the question using only the context below.

Context:
{context}

Question:
{question}
"""

response = model.invoke(prompt)
print("\nFinal Answer:")
print(response.content)
```
